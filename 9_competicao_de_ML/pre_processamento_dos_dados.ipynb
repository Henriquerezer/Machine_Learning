tratamento dos dados 

```python
# Dummificar está coluna. junto com feriado e dia

variaveis_categoricas = ['estação', 'dia', 'feriado']
dados = pd.get_dummies(dados, columns = variaveis_categoricas, drop_first = True) 
dados.head(5)
```

```python
# categorização da variavel chuva / # acho que vou separar os dados, em duas variaveis categóricas, 0 ( sem chuva) e 1 ( com chuva ) , pois os valores de 0 são muito mais comuns

def chuva (x):
  if x == 0 :
    return 0
  else:
    return 1

dados['chuva'] = dados['chuva'].apply(chuva)
dados.head(5)
```

```python
#normalização de variaveis continuas, ( as varaiveis ficaram entre 0 e 1)

# temperatura 
dados['temperatura'] = dados['temperatura'] / 100

# umidade
dados['umidade'] = dados['umidade'] / 100

#sol 
dados['sol'] = dados['sol'] / 10

#vento
dados['vento'] = dados['vento'] / 10

# hora
dados['hora'] = dados['hora'] / 100

dados.head(5)
```

```python
# tratamento da variavel visibilidade 

dados['visibilidade'] = dados['visibilidade'].replace({'%': ''}, regex=True)
dados.head()
```

```python
dados['visibilidade'] = dados['visibilidade'].astype(float, errors = 'raise')
```

```python
dados['visibilidade'].value_counts()
```

```python
def visi (x):
  if x <= 30:
    return 'Umidade de 0% a 30%'
  elif x <= 60:
    return 'Umidade de 30% a 60%'
  elif x <= 95:
    return 'Umidade de 60% a 95%'
  else:
    return 'Umidade +95%'

dados['visibilidade'] = dados['visibilidade'].apply(visi)

dados.head()
```

```python
dados['visibilidade'].value_counts().plot(kind = 'pie');
```

```python
categorica = ['visibilidade']
dados = pd.get_dummies(dados, columns = categorica, drop_first = True )
dados.head(5)
```
