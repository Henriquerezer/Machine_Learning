**importações principais (basicas)**

```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt
```

**inserindo a base de dados a ser utilizada texto em itálico**

```python
url = 'https://raw.githubusercontent.com/gitflai/DDS4/main/insurance.csv'
dados = pd.read_csv(url)
```

# ***MISE UN PLACE DOS DADOS***

```python
#TRANSFORMANDO OS DADOS DA COLUNA "SMOKER" ( YES PARA 1) E (NO PARA 0)
dados['smoker'] = dados['smoker'].map({'yes' : 1, 'no' : 0})

#FILTRANDO SOMENTE OS DADOS DE INTERESSE 
dados = dados[['bmi', 'smoker', 'charges']]


# separação de variáveis preditoras e variával resposta
X = dados[['bmi', 'charges']]
y = dados['smoker'] 
```

**Optei pela técnica de validação cruzada stratificada **

```python
# Configuração da Validação Cruzada

from sklearn.model_selection import cross_validate
from sklearn.model_selection import RepeatedStratifiedKFold

validacao = RepeatedStratifiedKFold(n_splits = 10, n_repeats = 100)
```

```python
#simulando apenas um modelo 
modelo1 = LogisticRegression()
medidas1 = cross_validate(modelo1, X, y, cv = validacao, scoring = 'accuracy')
round(medidas1['test_score'].mean(),4)
```

# melhor modelo de regressão para a variável **charges**, utilizando a métrica **EQM**



MISE EN PLACE 

```python
dados['charges'] = dados['charges']/1000 
X1 = dados[['bmi', 'smoker']]
y2 = dados['charges'] 
```

```python
from lazypredict.Supervised import LazyRegressor

from sklearn.utils import shuffle
```

```python
X1 = X1.astype(np.float32)
offset = int(X1.shape[0] * 0.9)


X_train, y_train = X1[:offset], y2[:offset]
X_test, y_test = X1[offset:], y2[offset:]

reg = LazyRegressor(verbose=0, ignore_warnings=False, custom_metric=None)
models3, predictions = reg.fit(X_train, X_test, y_train, y_test)

models3
```
